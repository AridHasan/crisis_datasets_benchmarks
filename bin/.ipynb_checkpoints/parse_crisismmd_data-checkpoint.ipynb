```python
file_name="/Users/firojalam/QCRI/exp_text_classification_bert/data/crisismmd/task_humanitarian_text_img_train.tsv"
delim="\t"
humanitarian_df = pd.read_csv(file_name,sep=delim);
#print(humanitarian_df.head())
```

```python
class Instance(object):
    def __init__(self,id=1,tweet_text="",label_text=""):
        self.id = id
        self.tweet_text = tweet_text
        self.label_text = label_text
        
humanitarian_dict={}
for index, row in humanitarian_df.iterrows():
    tweet_id=str(row['tweet_id'])
    tweet_text=row['tweet_text']    
    label_text=row['label_text']
    if(label_text=="vehicle_damage"):
        label_text="infrastructure_and_utility_damage"
        
    inst = Instance(id=tweet_id,tweet_text=tweet_text,label_text=label_text)
    humanitarian_dict[tweet_id]=inst
print(len(humanitarian_dict))
```

```python
file_name="/Users/firojalam/QCRI/exp_text_classification_bert/data/crisismmd/task_informative_text_img_train.tsv"
delim="\t"
informative_df = pd.read_csv(file_name,sep=delim);

informative_dict={}
for index, row in informative_df.iterrows():
    tweet_id=str(row['tweet_id'])
    tweet_text=row['tweet_text']    
    label_text=row['label_text']
    inst = Instance(id=tweet_id,tweet_text=tweet_text,label_text=label_text)
    informative_dict[tweet_id]=inst
print(len(informative_dict))
```
